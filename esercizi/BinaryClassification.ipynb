```python
#ci sono righe e due colonne, dividiamo il training set e il test set in due gruppi
X_train, x_test, y_train, y_test = train_test_split(df[["age"]], df.bought_insurance, test_size=0.1)


```

```python
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(x_test)
```

```python
#se noti, gli esiti hanno senso dato che la persona di 61 anni compra l'assicurazione, quelle + giovani no
model.score(x_test,y_test)
```
