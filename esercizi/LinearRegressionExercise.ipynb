```python
#esercizio che prevede di effettuare una previsione dei costi nel 2020

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(".\\Bicocca\\Stage\\canada_per_capita_income.csv")
print(df.columns)
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capital income")
plt.scatter(df.year, df['per capita income (US$)'], color = "blue")
```

```python
rm = linear_model.LinearRegression()
rm.fit(df[["year"]], df['per capita income (US$)'])
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capital income")
plt.scatter(df.year, df['per capita income (US$)'], color = "blue")
plt.plot(df["year"], rm.predict(df[["year"]]))
plt.plot(2020, 41288.69409442)
```
